```python
import numpy as np #numpy는 수학 술식을 많이 포함
import matplotlib.pyplot as plt
```

```python
x=np.linspace(0,14,100) #0부터 14까지 랜덤 100개 생성 by list!!!
y=np.sin(x) #x이용 y값 100
plt.figure(figsize=(8,6)) #그래프 크기 조정
plt.plot(x,y, label="sin(x)") #matplotlib이용 line plot그래프 그리기 &label설정
plt.legend() #라벨 hue로 실행!
plt.title("sin plot") #제목 선정
```

```python
import seaborn as sns
tips=sns.load_dataset("tips") #seaborn 데이터 라이브러리에서 tips를 불러오기
#tip.head() #앞5개의 리스트 불러오기
# tip.tail() 뒤 5개 리스트 불러오기
plt.figure(figsize=(8,6)) #matplot으로 그래프 사이즈 조정 &호출
sns.barplot(x="day", y="total_bill",hue="sex", data=tips) #seaborn은 plt안에 존재하니까 sms이용해 그래프 정보 불러오기 & 그리기
plt.title("Title place") #legend와 다른점은 hue는 그래프 구성에 영향있음!
plt.show #정보 그리기
```

```python
sns.set_style("darkgrid") # 종류 : darkgrid, whitegrid, dark, white, ticks 배경 선택!
sns.lmplot(x="total_bill", y="tip", data=tips, height = 6) #그래프 사이즈 조정
# plt.show() #sns는 스스로 호출 가능!
```

```python
print(tips.shape) #데이터 행과 열
print(tips.columns)  #열의 종류(제목열) 그리기
print(tips.describe()) #데이터 요약
```

```python
fg=sns.load_dataset("flights")
sns.scatterplot(x="year", y="passengers",hue="month", data=fg) #hue는 색으로 구분 
```

```python
plt.figure(figsize=(12,8))
sns.boxplot(x="month", y="passengers", data=fg) #height=8 이 그래프는 
```

```python
fgpivot=fg.pivot("month","year","passengers") #pandas랑 비슷
fgpivot
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(fgpivot, annot=True, annot_kws={'size':15} ,fmt="d",cmap="Reds")
#fmt는 heatmap속 글형식 d=실수, annot은 값을 출력 여부 , cmap은 색 지정 ,annot_kws은 내부 값 조정
```
